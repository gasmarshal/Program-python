```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tebakan = pd.read_csv('/content/predict_naive_bayes.csv')
tebakan
```

```python
outlook = tebakan.groupby(['outlook', 'play']).size()
temp = tebakan.groupby(['temp', 'play']).size()
humidity = tebakan.groupby(['humidity', 'play']).size()
windy = tebakan.groupby(['windy', 'play']).size()
play = tebakan.play.value_counts()
```

```python
print(temp)
print('------------------')
print(humidity)
print('------------------')
print(windy)
print('------------------')
print(outlook)
print('------------------')
print('play')
print(play)
print('\n======|Hasilnya adalah jeng jeng jeng|======\n')
#HASIL AKHIR ADALAH
p_x_yes = ((4/9)*(4/9)*(6/9)*(3/9))
p_x_no = ((0/5)*(2/5)*(1/5)*(3/5))

print('NILAI DARI YES ADALAH =', '%.3f'%p_x_yes)
print('NILAI DARI NO ADALAH =', '%.3f'%p_x_no)
if p_x_yes > p_x_no:
  print('Pilihan Play Adalah Yes')
else:
  print('Pilihan Play adalah No')
```
